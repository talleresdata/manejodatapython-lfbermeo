```python
import pandas as pd
import sys
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
#from pandas.tools.plotting import radviz
%matplotlib inline
```

```python
dataFrame = pd.read_json('data/log-course2.json')
```

```python
len(dataFrame['ip'].unique())
```

```python
dataFrame2 = dataFrame
dataFrame = dataFrame[['ip', 'event_source']].groupby(['ip'],as_index=False).count()
dataFrame
```

```python
dataFrame=dataFrame.sample(70)
```

## Gráfica tomando en cuenta las dirección ip de los usuarios, tomando una muestra de 70

```python
plt.figure(figsize=(10,15),dpi=100)
plt.barh(dataFrame['ip'].values,dataFrame['event_source'].values)
plt.title("Frecuencia de aparición de ip")
plt.xlabel("Frecuencia")
plt.ylabel("Ip")
```
